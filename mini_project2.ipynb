```python
import pymysql as m
import pandas as pd
```

```python
df1 = pd.read_csv("C:/mysql/news11.csv")
df1
```

```python
con = m.connect(host='localhost', user='root', password='0000', db='news', charset='utf8')
cur = con.cursor()
for i in range(len(df1)):
    q = 'insert into news values(%s, %s);'
    cur.execute(q, (str(df1["Rating"][i]), str(df1["review_text"][i])))
con.commit()
con.close()
```
